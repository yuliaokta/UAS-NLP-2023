```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca dataset
df = pd.read_csv('all_brands_kosmetik.csv')
```

```python
# Menampilkan informasi dataset
print(df.info())
```

```python
# Menampilkan statistik deskriptif
print(df.describe())
```

```python
# Menampilkan distribusi harga
sns.histplot(df['price_range'])
plt.xlabel('Price Range')
plt.ylabel('Count')
plt.title('Distribution of Price Range')
plt.show()
```

```python
# Menampilkan hubungan antara rating dan jumlah ulasan
plt.figure(figsize=(8, 6))
sns.scatterplot(x='average_rating', y='total_reviews', data=df, color='green', alpha=0.7, edgecolor='black', linewidth=0.5)  # Menambahkan garis tepi pada titik
plt.xlabel('Average Rating')
plt.ylabel('Total Reviews')
plt.title('Relationship between Average Rating and Total Reviews')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Menampilkan korelasi antara variabel numerik
numeric_cols = ['beauty_point_earned', 'average_rating', 'total_reviews']
plt.figure(figsize=(8, 6))
sns.heatmap(df[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)  # Menambahkan ketebalan garis antar sel
plt.title('Correlation between Numeric Variables')
plt.show()
```

```python
# Menampilkan jumlah produk per brand
brand_counts = df['brand_name'].value_counts()
plt.figure(figsize=(10, 6))
sns.barplot(x=brand_counts.index, y=brand_counts.values)
plt.xlabel('Brand')
plt.ylabel('Count')
plt.title('Number of Products per Brand')
plt.xticks(rotation=90)
plt.show()
```
