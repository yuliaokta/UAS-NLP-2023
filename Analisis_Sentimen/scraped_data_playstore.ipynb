```python
# Import library yang dibutuhkan
from google_play_scraper import app

import pandas as pd

import numpy as np
```

```python
#mengambil data sebagian
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.tiktokshop.seller',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=17000,
    filter_score_with=None
)

```

```python
df_busu = pd.DataFrame(np.array(result),columns=['reviews'])

df_busu = df_busu.join(pd.DataFrame(df_busu.pop('reviews').tolist()))

df_busu.head()
```
