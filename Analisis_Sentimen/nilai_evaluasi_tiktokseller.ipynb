```python
import matplotlib.pyplot as plt

# Nilai evaluasi
accuracy = 0.90
precision = 0.93
recall = 0.91

# Mengubah nilai evaluasi menjadi persen
accuracy_percent = accuracy * 100
precision_percent = precision * 100
recall_percent = recall * 100

# Membuat label untuk setiap nilai evaluasi
labels = ['Accuracy', 'Precision', 'Recall']
values = [accuracy_percent, precision_percent, recall_percent]

# Membuat plot
plt.bar(labels, values)
plt.ylim(0, 100)  # Mengatur batas nilai y dari 0 hingga 100
plt.xlabel('Evaluasi')  # Memberi label pada sumbu x
plt.ylabel('Nilai (%)')  # Memberi label pada sumbu y
plt.title('Visualisasi Nilai Evaluasi')  # Memberi judul pada plot

# Menampilkan nilai evaluasi di atas setiap batang
for i in range(len(labels)):
    plt.text(i, values[i], str(round(values[i], 2)) + '%', ha='center', va='bottom')

# Menampilkan plot
plt.show()
```
